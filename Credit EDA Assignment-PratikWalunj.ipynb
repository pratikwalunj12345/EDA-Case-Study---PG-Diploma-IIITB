```python
#importing the application_data.csv dataset

app_data = pd.read_csv('application_data.csv')
app_data.head()
```

#### Here I observed that 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH' & 'DAYS_LAST_PHONE_CHANGE' columns have negative values, which is not not practically possible. Hence I converted all the negative values into positive and divide these values by 365 to convert days into year.

```python
days_col = ['DAYS_BIRTH' ,'DAYS_EMPLOYED' ,'DAYS_REGISTRATION' ,'DAYS_ID_PUBLISH','DAYS_LAST_PHONE_CHANGE']
app_data[days_col] = app_data[days_col].abs()
app_data[days_col] = app_data[days_col]/365
app_data[days_col].describe()
```

```python
# Cleaning the missing data

# listing the null values columns having more than 30%

col_empty=app_data.isnull().sum()
col_empty=col_empty[col_empty.values>(0.3*len(col_empty))]
len(col_empty)
```

Here we come to know that there are 64 columns having null values greater than 30% in the dataset

```python
# Removing those 64 columns
col_empty = list(col_empty[col_empty.values>=0.3].index)
app_data.drop(labels=col_empty,axis=1,inplace=True)
print(len(col_empty))
```

```python
# Checking the columns having less null percentage

app_data.isnull().sum()/len(app_data)*100
```

Here we come to know that 'AMT_ANNUITY' column is having very few null values rows.So I tried to impute the missing values



This column have an outlier which is very large, hence it is not correct to fill those missing values with mean. So, I used Median to fill those missing values with median value

```python
# Filling missing values with median

val=app_data['AMT_ANNUITY'].median()

app_data.loc[app_data['AMT_ANNUITY'].isnull(),'AMT_ANNUITY']=val
```

```python
# Searching for the column for null values

app_data.isnull().sum()
```

Now, almost all necessary columns have been with zero null values

```python
# Removing rows having null values greater than or equal to 30%

row_empty=app_data.isnull().sum(axis=1)
row_empty=list(row_empty[row_empty.values>=0.3*len(app_data)].index)
app_data.drop(labels=row_empty,axis=0,inplace=True)
print(len(row_empty))
```

```python
# We will remove unwanted columns from this dataset

unwanted_col=['FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE',
       'FLAG_PHONE', 'FLAG_EMAIL','REGION_RATING_CLIENT','REGION_RATING_CLIENT_W_CITY','FLAG_EMAIL','CNT_FAM_MEMBERS', 'REGION_RATING_CLIENT',
       'REGION_RATING_CLIENT_W_CITY','DAYS_LAST_PHONE_CHANGE', 'FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_3','FLAG_DOCUMENT_4', 'FLAG_DOCUMENT_5', 'FLAG_DOCUMENT_6',
       'FLAG_DOCUMENT_7', 'FLAG_DOCUMENT_8', 'FLAG_DOCUMENT_9','FLAG_DOCUMENT_10', 'FLAG_DOCUMENT_11', 'FLAG_DOCUMENT_12',
       'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14', 'FLAG_DOCUMENT_15','FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17', 'FLAG_DOCUMENT_18',
       'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20', 'FLAG_DOCUMENT_21']

app_data.drop(labels=unwanted_col,axis=1,inplace=True)
```

```python
#Checking type of Gender present
app_data["CODE_GENDER"].value_counts()
```

```python
# Count plotting in logarithmic scale

def plots(app_data,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 15
    plt.rcParams['axes.titlesize'] = 17
    plt.rcParams['axes.titlepad'] = 21
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(app_data[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = app_data, x= col, order=app_data[col].value_counts().index,hue = hue,palette='mako') 
        
    plt.show()
```

```python
# Now, plotting the above correlation with heat map as it is the best choice to visulaize

# figure size

def tar_corr(data,title):
    plt.figure(figsize=(25,25))
    plt.rcParams['axes.titlesize'] = 20
    plt.rcParams['axes.titlepad'] = 50

# heatmap with a color map of choice

    
    sns.heatmap(data, cmap="YlOrRd",annot=True,vmin= -1,vmax=1)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```

```python
# Here I have Cleaned the missing data

# listing the null values columns having more than 30%

empty_col1=prev_data.isnull().sum()
empty_col1=empty_col1[empty_col1.values>(0.3*len(empty_col1))]
len(empty_col1)
```

```python
# here removing the above 15 columns

empty_col1 = list(empty_col1[empty_col1.values>=0.3].index)
prev_data.drop(labels=empty_col1,axis=1,inplace=True)
prev_data.shape
```
